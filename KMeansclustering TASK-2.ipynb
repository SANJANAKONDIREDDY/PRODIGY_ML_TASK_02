```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
import pandas as pd

# Assuming X is your dataset (e.g., standardized features)
X = pd.DataFrame({
    'Age': np.random.rand(100),
    'Annual Income (k$)': np.random.rand(100)
})

fig, ax = plt.subplots(2, 2, figsize=(15, 8))
cluster_counts = [2, 3, 4, 5]

for idx, n_clusters in enumerate(cluster_counts):
    # Create KMeans instance for the current number of clusters
    km = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=100, random_state=42)
    
    # Determine the position of the subplot
    row = idx // 2
    col = idx % 2
    
    # Create SilhouetteVisualizer instance with KMeans instance
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[row, col])
    
    # Fit the visualizer
    visualizer.fit(X)
    
    # Set the title for each subplot
    ax[row, col].set_title(f'KMeans with {n_clusters} Clusters')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Sample DataFrame with standardized data (assuming dfa_std is standardized)
dfa_std = np.random.rand(100, 2)  # Replace this with your actual standardized data

# Fit the KMeans model
kmeans = KMeans(n_clusters=4, random_state=42).fit(dfa_std)
labels = kmeans.labels_

# Create a new DataFrame with the standardized data and the cluster labels
new_dfa = pd.DataFrame(data=dfa_std, columns=['Age', 'Annual Income (k$)'])
new_dfa['label_kmeans'] = labels

# Create the scatter plot
fig, ax = plt.subplots(figsize=(10, 7))

# Define colors and edgecolors for each cluster
colors = ['pink', 'black', 'brown', 'purple']
edgecolors = ['blue', 'blue', 'blue', 'blue']

# Plot the clusters
for i, color, edgecolor in zip(range(4), colors, edgecolors):
    ax.scatter(new_dfa["Annual Income (k$)"][new_dfa["label_kmeans"] == i], 
               new_dfa["Age"][new_dfa["label_kmeans"] == i],
               color=color, s=100, edgecolor=edgecolor, label=f'Cluster {i+1}')

# Plot cluster centers
centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], c='yellow', s=500, alpha=0.75, label='Centroids', marker='X')

# Label axes and the plot
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Age')
ax.set_title('KMeans Clustering of Age vs Annual Income')

# Add legend
ax.legend()

# Show plot
plt.show()
```
